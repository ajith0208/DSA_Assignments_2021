```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the dataset to the python environment.

```python
car = pd.read_csv('cars_data.csv')
car
```

# 2. Check for the null values present in the dataset.

```python
car.isna().sum()
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(15,12))
sns.countplot(x='Buyer Gender',data=car)
plt.title('Male vs Female',fontsize=20,fontweight='bold')
plt.grid(True)
```

Count of female is slightly higher than male were participated in sales
